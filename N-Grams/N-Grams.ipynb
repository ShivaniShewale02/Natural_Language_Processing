# N Grams

- N-grams are one of the most fundamental techniques in NLP, especially in classical NLP methods used before deep learning models became dominant. They are widely used in language modeling, text prediction, search engines, spelling correction, and text classification.
- N-grams are sequences of N consecutive words used to capture contextual relationships between words in NLP tasks.

### What is N Grams?
- An N-gram is a sequence of N consecutive words (or tokens) from a text.
1. N --> number of tokens
2. Gram --> sequence

So an N-Gram = N Consecutice Tokens

Example;
```
"I love learning natural language processing"
```
| N | Name      | Example                    |
| - | --------- | -------------------------- |
| 1 | Unigram   | I, love, learning, natural |
| 2 | Bigram    | I love, love learning      |
| 3 | Trigram   | I love learning            |
| 4 | Four-gram | I love learning natural    |

---

# Code Implementation

````python
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

text = "I love machine learning and natural language processing"

tokens = word_tokenize(text)

bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

print("Bigrams:", bigrams)
print("Trigrams:", trigrams)
````
