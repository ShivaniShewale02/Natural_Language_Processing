# Natural Language Processing (NLP)

- Natural Language Processing (NLP) is a branch of AI that enables computers to understand, interpret, generate, and interact with human language.
- Natural Language Processing (NLP) a feild of Artificial Intelligence (AI) which is used to narrow the communication gap in between the computer and human.
- It enables computer to understand, interpret and generate human understandable content (Human Readable language like English).
- NLP converts human language into numerical representations so machines can understand, analyze, and generate text intelligently.

It sits at the intersection of:
1. Mathematics
2. Machine Learning
3. Deep Learning
4. Linguistics

### Two Main Components Of NLP

##### 1. Natural Language Understanding (NLU)
- NLU helps machine understand and interpret human language, extracting meaning from the text. Basically helps in understanding human context and convert it into data.
- Key process of NLU is Tokenization, POS, Parsing, NER, etc.

##### 2. Natural Language Generation (NLG)
- NLG enables machines to generate human readable text or speech, creating meaningfull and grammatically correct content.
- Key Process of NLG is Content planning, sentence formation, text formating, etc.

### NLP Pipeline (Complete Flow)

```
Raw Text
   ↓
Text Cleaning
   ↓
Tokenization
   ↓
Stopword Removal
   ↓
Stemming/Lemmatization
   ↓
Vectorization (BoW / TF-IDF / Embeddings)
   ↓
Model (ML/DL)
   ↓
Prediction / Generation
```
