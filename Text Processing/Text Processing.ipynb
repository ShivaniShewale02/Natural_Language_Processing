# Text Processing

- Text Processing = Converting raw, messy human text into structured, meaningful, machine-readable format.

Example;
``` 
Raw Text:
"Hey!!! I'm Devashish, working on AI & ML projects since 2023 😊"

Processed Text:
["devashish", "work", "ai", "ml", "project", "2023"]
```

### Complete Text Processing Pipeline

```
Raw Text
   ↓
Lowercasing
   ↓
Remove Noise (HTML, URLs, Emojis)
   ↓
Tokenization
   ↓
Remove Stopwords
   ↓
Stemming / Lemmatization
   ↓
Handling Negations
   ↓
Normalization
   ↓
Vectorization
```
- Sometimes depending on the requirnment we do use more than this apporaches for to work on processing part for text.
