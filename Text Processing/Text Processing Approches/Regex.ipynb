# Regex

- Regular Expressions (Regex) are patterns used to match, search, extract, and manipulate text.
- They are widely used in Natural Language Processing (NLP), text processing, data cleaning, validation, and information extraction.
- Regex is a text pattern language used to find specific strings inside large text.
- Regular Expressions (Regex) are sequences of characters used to define search patterns for matching, extracting, and manipulating text.

Example;
```
Text: "My email is dev@gmail.com"
Regex Pattern: "[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"
Result: "dev@gmail.com"
```

## Metacharacters (Special Regex Symbols)
- Regex contains special characters with meanings which are called metacharacters.

| Symbol | Meaning           |
| ------ | ----------------- |
| .      | any character     |
| ^      | start of string   |
| $      | end of string     |
| *      | zero or more      |
| +      | one or more       |
| ?      | optional          |
| {}     | exact repetitions |
| []     | character set     |
| \      | escape character  |
| ()     | grouping          |


---

# Code Example

````python
import re

text = "Contact me at dev@gmail.com or call 9876543210."

email = re.findall(r"\w+@\w+\.\w+", text)
phone = re.findall(r"\d{10}", text)

print(email)
print(phone)
````

````python
import re

text = "I love NLP and AI"

match = re.search("NLP", text)

print(match)
````
