# Lemmatization

- Lemmatization is a text normalization technique that reduces words to their base or dictionary form called a lemma.
- Unlike stemming, lemmatization uses linguistic knowledge such as vocabulary and grammar (Part-of-Speech) to convert words into valid dictionary words.
- Lemmatization is a text normalization technique in NLP that converts words into their base dictionary form (lemma) using vocabulary and morphological analysis.
- Lemmatization reduces words to their dictionary form (Lemma) by considering th context & part of speach. It provies accurate & meaningfull words.

Example;
| Word    | Lemma |
| ------- | ----- |
| running | run   |
| better  | good  |
| studies | study |
| was     | be    |
| cars    | car   |

- So instead of producing artificial roots like studi, lemmatization produces actual meaningful base words.

---- 

## Code

```python
# Using NLTK
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
words = ["running", "studies", "better", "cars"]

for word in words:
    print(word, "->", lemmatizer.lemmatize(word))
```

```python
# Using SpaCy

import spacy

nlp = spacy.load("en_core_web_sm")
text = "The striped bats are hanging on their feet"
doc = nlp(text)

for token in doc:
    print(token.text, "->", token.lemma_)
```
