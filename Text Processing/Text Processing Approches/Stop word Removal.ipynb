# Stop-Word Removal

- Stopwords are commonly occurring words in a language that carry little semantic meaning and are often removed during text processing.
- Stopword removal is a preprocessing technique in NLP where commonly occurring words with little semantic meaning are removed from text to reduce noise and improve model performance.
- Stopword removal is while we process text in NLP, raw text contains unnecessory words that do not contribute much to the meaning of a sentence is being removed by stopword removal approach, such words as "is", "the", "and", etc.
- Removing stopword from the sentence or text is stopword removal approach/process.

### Types of Stopwords
- Stopwords can be categorized into four major groups.



#### 1. Standard Stopword
- These are commonly used function words.

Example: the, is, at, which, on

#### 2. Domain-Specific Stopwords
- Certain words are frequent in specific domains.

Example: study, paper, result, method

#### 3. Contextual Stopwords
- Words that are frequent in a specific dataset.



#### 4. Numerical Stopwords
- Numbers that don't contribute to meaning.

Example: 2021, 123, 456

# Stopword Removal in NLP Libraries

```python
# Using NLTK
import nltk 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download("stopwords")

text = "This is an example of Stopword removal in NLP"

words = word_tokenize(text)
filtered_words = [word for word in words if word.lower() not in stopwords.words("english")]

print(words)
print(filtered_words)


## from nltk.corpus import stopwords
## stop_words = stopwords.words('english')
```
