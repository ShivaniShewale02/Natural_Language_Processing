# Text Normalization

- Text Normalization is one of the most important preprocessing steps in NLP. It converts text into a standardized and consistent format so that machines can process it correctly.
- Text normalization is the process of converting text into a standard, consistent form by removing variations such as case differences, punctuation, spelling errors, and morphological variations to improve the performance of NLP models.
- Text Normalization is a process of transforming text into a canonical (standard) form by removing variations that do not affect meaning.
- Normalization helps reduce vocabulary size, noise, and ambiguity.

Example;
```
Original Text: "I can't believe AI is sooo cooool!!! 😃"
Normalized Text: "i cannot believe ai is so cool"
```

### 1. Case Normalization (Lowercasing)

- Convert all text to lowercase.

Example;
```
Machine Learning
machine learning
MACHINE LEARNING
```

Normalized;
```
machine learning
```

##### Advantages:

- Reduces vocabulary size
- Simple and fast

##### Limitation:
- Not suitable for Named Entity Recognition


##### Example;
apple --> fruit
Apple --> company

# Code

````python
# Python 
text = "Hello, I Am Devashish Sharma"
text_normalized = text.lower()
text_normalized
````

----

### 2. Punctuation Normalization

- Remove unnecessary punctuation.

Example:
" Hello!!! How are you??? "

Normalized:
" hello how are you "

## Code

````python
import string

text = "Hello!!! How are you???"
text_lower = text.lower()
text_normalized = text_lower.translate(str.maketrans('', '', string.punctuation))
text_normalized
````

---

### 3. Whitespaces Normalization

- Remove extra spaces.



## Code

````python
import re 

text = "I     Love     AI"
text_lower = text.lower()
text_normalized = text = re.sub(r'\s+', ' ', text)
text_normalized
````

---

### 4. Constraction Expansion

- Expand Shortened forms.

Example:
| Contraction | Expanded |
| ----------- | -------- |
| can't       | cannot   |
| don't       | do not   |
| I'm         | I am     |
| it's        | it is    |

## Code

````python
contractions = {
"can't":"cannot",
"don't":"do not",
"i'm":"i am"
}

text = "I'm Devashish Sharma, I don't like reading and I can't force myself to do what I don't like."
text_normalized = text.lower()

for k, v in contractions.items():
    text_normalized = text_normalized.replace(k, v)

print(text_normalized)


import contractions

text = "I'm Devashish Sharma, I don't like reading and I can't force myself."
expanded = contractions.fix(text)

print(expanded)
````

### 7. Repeated Character Normalization

- Social Media contains repeated characters.
  


## Code

````python
import re 

text = 'sooo happpyyyy'
text_normalized = re.sub(r'(.)\1+', r'\1\1', text)
print(text_normalized)
````

---

### 8. Number Normalization

- Number can be normalized in multiple ways.



## Code

````python
import re

text = "I have 3 apples"
text_normalized = re.sub(r'\d+', 'three', text)
text_normalized
````

### Unicode Normalization

- Text from the web often has multiple encodings.



- Unicode normalization ensures consistent encoding.

Forms:

| Form | Description                 |
| ---- | --------------------------- |
| NFC  | canonical composition       |
| NFD  | canonical decomposition     |
| NFKC | compatibility composition   |
| NFKD | compatibility decomposition |



## Code

````python
import unicodedata

text = "café"
text_normalized = unicodedata.normalize("NFKC", text)
text_normalized
````
