# Tokenization

- Tokenization is the process of splitting/breaking text into smaller units called tokens.
- Tokenization is the process of breaking text into smaller units called tokens, such as words, sentences, or subwords, which serve as the basic units for NLP models.
- These tokens can be works, sentences or subwords, depending on the level of tokenization applied.
- It is a foundation step of NLP as it prepairs the text for further processing.

Example;
```
Input:
"I love Machine Learning"

Tokens:
["I", "love", "Machine", "Learning"]
```

# Types of Tokenization

1. Sentence Tokenization
2. Word Tokenization
3. Charactor Tokenization
4. Subword Tokenization
5. Byte-level Tokenization

---

### Sentence Tokenization

- Sentence Tokenization is used to split a large text into individual sentences, which is usefull for task like summarization, machine translation, sentimental analysis, etc.
- Sentence tokenization splits text into sentences.

Example;
```
Input:
"I love AI. It is amazing."

Output:
["I love AI.", "It is amazing."]
```

#### Code

````python
# Using nltk
import nltk
from nltk.tokenize import sent_tokenize

text = "I love AI. It is amazing."
sentences = sent_tokenize(text)

print(sentences)
````

````python
# Using SpaCy
import spacy
nlp = spacy.load("en_core_web_sm")

text = "I love AI. It is amazing."
doc = nlp(text)
sentence_token = [sent.text for sent in doc.sents]

print(sentence_token)
````

---

### Word Tokenization

- Word Tokenization splits sencentences into words.
- Word Tokenization devides a sentence into individual words or subwords, this is essential for text analysis, speech recorgnition and search engine.

Example;
```
Input:
"I love NLP"

Output:
["I", "love", "NLP"]
```

#### Code

````python
# Using NLTK
from nltk.tokenize import word_tokenize

text = "I love NLP"
tokens = word_tokenize(text)

print(tokens)
````

````python
# Using SpaCy
import spacy
nlp = spacy.load("en_core_web_sm")

text = "I love AI. It is amazing."
doc = nlp(text)
word_token = [token.text for token in doc]

print(word_token)
````

---

### Character Tokenization

- Breaks text into individual characters.

```
Input:
AI

Output:
['A', 'I']
```

#### Code

````python
text = "AI"
tokens = list(text)

print(tokens)
````
