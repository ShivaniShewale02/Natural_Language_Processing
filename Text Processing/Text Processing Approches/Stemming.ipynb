# Stemming

- Stemming is a text normalization technique used in Natural Language Processing (NLP) to reduce words to their root form (stem) by removing suffixes or prefixes.
- The stem may not always be a valid dictionary word, but it represents the base meaning.
- Stemming is a text normalization technique in NLP that reduces words to their root form by removing suffixes and prefixes.
- Stemming is used t remove prefixes & suffixes to obtain the root form of a word to group words with a similar basic meaning together.
- Its accuracy is low as it is preffered when the meaning of word in not immportant.
- It usually operates on single word without knowing the knowledge of the content and it is faster as it only removes prefix/suffix.

Example:

| Original Word | Stem          |
| ------------- | ------------- |
| playing       | play          |
| played        | play          |
| player        | player / play |
| studies       | studi         |
| studying      | studi         |
| better        | bet           |

### Where Stemming is used

1. Search Engines
2. Information Retrieval
3. Text Classification
4. Sentiment Analysis
5. Topic Modeling

etc

---

## Code

```python
# Using NLTK (Porter Stemmer Example)

import nltk 
from nltk.stem import PorterStemmer

ps = PorterStemmer()
words = ["playing","played","player","studies","studying"]

for w in words:
    print(w, "->", ps.stem(w))
```

```python
# Using NLTK (Snowball Stemmer Example)

from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("english")
words = ["running","studies","playing"]

for w in words:
    print(w, "->", stemmer.stem(w))
```

```python
# Using NLTK (Lancaster Stemmer Example)

from nltk.stem import LancasterStemmer

stemmer = LancasterStemmer()
words = ["running","studies","playing"]

for w in words:
    print(w, "->", stemmer.stem(w))
```
