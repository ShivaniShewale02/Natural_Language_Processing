# Part Of Speech (POS) Tagging


- Part-of-Speech (POS) Tagging is a Natural Language Processing (NLP) technique used to assign grammatical categories (tags) to each word in a sentence.
- These categories describe how the word functions in the sentence.
- Part-of-Speech tagging is the process of assigning grammatical categories such as noun, verb, adjective, etc., to each word in a sentence based on its context.

Example;
```
"The quick brown fox jumps over the lazy dog."
```

POS Tagging Results:
| Word  | POS Tag          |
| ----- | ---------------- |
| The   | Determiner (DT)  |
| quick | Adjective (JJ)   |
| brown | Adjective (JJ)   |
| fox   | Noun (NN)        |
| jumps | Verb (VBZ)       |
| over  | Preposition (IN) |
| the   | Determiner (DT)  |
| lazy  | Adjective (JJ)   |
| dog   | Noun (NN)        |


### 1. What is a Part of Speech?
- A Part of Speech is a grammatical category that describes the role of a word in a sentence.

Main categories in English grammar:
| POS         | Meaning            | Example   |
| ----------- | ------------------ | --------- |
| Noun        | Person/place/thing | dog, city |
| Verb        | Action/state       | run, eat  |
| Adjective   | Describes noun     | big, fast |
| Adverb      | Describes verb     | quickly   |
| Pronoun     | Replaces noun      | he, she   |
| Preposition | Shows relation     | in, on    |
| Conjunction | Connects words     | and, but  |
| Determiner  | Defines noun       | the, a    |

## Code Example

````python
# Using NLTK
import nltk

# download required datasets
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

sentence = "The quick brown fox jumps over the lazy dog"
tokens = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(tokens)

print(pos_tags)
````

````python
# Using SpaCy
import spacy

nlp = spacy.load("en_core_web_sm")
text = "Elon Musk founded Tesla in 2003"
doc = nlp(text)

for token in doc:
    print(token.text, "-->",token.pos_)
````
