# Word2Vec

- Word2Vec is an algorithm that converts words into dense numerical vectors (embeddings).
- Word2Vec is a neural network based algorithm that learns dense vector representations of words by predicting context words from a target word or vice versa.
- Instead of representing words as large sparse vectors (like one-hot encoding), Word2Vec creates compact vectors that capture meaning.
- Developed by Google in 2013.



---

## Code

```python
from gensim.models import Word2Vec

sentences = [
    ["I","love","machine","learning"],
    ["machine","learning","is","powerful"],
    ["I","love","AI"]
]

model = Word2Vec(sentences, vector_size=100, window=2, min_count=1)

print(model.wv["machine"])
```

```python
model.wv.most_similar("machine")
```
